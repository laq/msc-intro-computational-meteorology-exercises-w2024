```python
import matplotlib.pyplot as plt
import numpy as np

from mpl_toolkits.axes_grid1.inset_locator import inset_axes
# Generate sample data
data = np.random.rand(10, 10)

# Create the plot
fig, ax = plt.subplots()
cax = ax.imshow(data, cmap='viridis')


inset_ax = inset_axes(ax, width="5%", height="30%", loc='upper right', borderpad=-2)


# Add the colorbar
cbar = fig.colorbar(cax, cax=inset_ax)
cbar.set_label('Intensity')

plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.animation
import numpy as np

# Generate time values
t = np.linspace(0, 2 * np.pi, 100)
x = np.sin(t)

# Create the figure and axis
fig, ax = plt.subplots()
ax.axis([0, 2 * np.pi, -1, 1])

# Create an initial scatter plot (empty)
sc = ax.scatter([], [])

# Animation function
def animate(i):
    sc.set_offsets(np.column_stack((t[:i], x[:i])))  # Update the positions of the scatter points

# Create animation
ani = matplotlib.animation.FuncAnimation(fig, animate, frames=len(t), interval=50)

# Display animation in Jupyter notebook
from IPython.display import HTML
HTML(ani.to_html5_video())
```
